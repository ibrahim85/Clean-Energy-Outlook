```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv('CA.csv')

#Data Preprocessing
data['HYTCP_scaled']=preprocessing.scale(data['HYTCP'])
data['GDP_scaled']=preprocessing.scale(data['GDP'])
data['CLPRB_scaled']=preprocessing.scale(data['CLPRB'])
data['EMFDB_scaled']=preprocessing.scale(data['EMFDB'])
data['ENPRP_scaled']=preprocessing.scale(data['ENPRP'])
data['PAPRB_scaled']=preprocessing.scale(data['PAPRB'])
data['NGMPB_scaled']=preprocessing.scale(data['NGMPB'])
data['PCP_scaled']=preprocessing.scale(data['PCP'])
data['ZNDX_scaled']=preprocessing.scale(data['ZNDX'])
data['OIL_PRICE_scaled']=preprocessing.scale(data['OIL_PRICE'])
```

```python
#Dividing data into training and testing data
training, testing = train_test_split(data, test_size = 0.1)
training=training.sort_values('Year')
testing=testing.sort_values('Year')

training_set_x=training[['HYTCP_scaled','GDP_scaled','CLPRB_scaled','EMFDB_scaled','ENPRP_scaled','PAPRB_scaled','NGMPB_scaled','PCP_scaled','ZNDX_scaled','OIL_PRICE_scaled']]
training_set_y=training['HYTCP_scaled']
testing_set_x=testing[['HYTCP_scaled','GDP_scaled','CLPRB_scaled','EMFDB_scaled','ENPRP_scaled','PAPRB_scaled','NGMPB_scaled','PCP_scaled','ZNDX_scaled','OIL_PRICE_scaled']]
testing_set_y=testing['HYTCP_scaled']

print(training_set_x)
print(training_set_y)
```

```python
#Ridge Regression
regr=linear_model.Ridge(alpha=0,normalize=True)
regr.fit(training_set_x,training_set_y)

fig=plt.figure(figsize=(10,4.5))
fig.suptitle('Ridge Regression', fontsize=14, fontweight='bold')

#Plot for training data
fig.add_subplot(121)
plt.plot(training['Year'],regr.predict(training_set_x),label='Predict',c='r')
plt.plot(training['Year'],training_set_y,label='Actual')
plt.legend(loc=0,fontsize=10)
plt.title('Training Data')
plt.xlabel('Year')
plt.ylabel('Scaled HYTCP')
plt.show()
```

```python
#Plot for testing data
fig.add_subplot(122)
plt.scatter(testing.Year,regr.predict(testing_set_x,marker='x',c='r',label='Predict')
plt.scatter(testing.Year,testing_set_y, marker='*',label='Actual')
a=list(testing.Year)
b=regr.predict(testing_set_x)
c=list(testing_set_y)
for i in range(len(a)):
    plt.plot([a[i],a[i]],[b[i],c[i]],'k--',linewidth=0.5,dashes=(3,3))
plt.legend(loc=2,fontsize=10)
plt.title('Testing Data')
plt.xlabel('Year')
plt.ylabel('Scaled HYTCP')

plt.tight_layout(pad=4, w_pad=4)
plt.show()
```
