```python
import pandas as pd
mydata_txt = pd.read_csv('climate_annual.txt',sep = ' ',encoding = 'utf-8')
for i in range(52):
    if i==44: 
        continue
    else: pass
    dforigin = pd.read_csv('../data of states/%s.csv' % (statelist[i]))
    dfstate=mydata_txt[mydata_txt.State==statesdic['%s' % statelist[i]]]
    dftoadd=dfstate[['Year','PCP','ZNDX']]
    dfnew=pd.merge(dforigin, dftoadd,on='Year',how='outer')
    dfnew.to_csv('../data of states/%s.csv'% (statelist[i]), encoding='utf-8', index=False)
    #if i==0: break
```
